```python
from wordcloud import WordCloud, STOPWORDS as stopwords
```

```python
text = open("Pride_and_Prejudice.txt","r").read()
text
```

```python
PnPwc = WordCloud(background_color = "white",
                  collocations = False, 
# 이때 collocations=True 이면 said King, said Hatter 같이 자주 나타나는 단어는 하나의 어구로 분류됨.
                  stopwords = stopwords)


# alice_novel이라는 텍스트 데이터를 사용하여 wordcloud를 generate하기
PnPwc.generate(text)
```

```python
PnPwc.to_image()
```

```python
PnPwc.to_file("PnP.png")
```

```python
PnPwc_2 = WordCloud(background_color = 'white',
                   collocations = False,
                   stopwords = stopwords,
                   width = 600, height = 400)

PnPwc_2.generate(text)
PnPwc_2.to_image()
```

```python
stopwords.update(["Mr","S"])
PnPwc_3 = WordCloud(background_color = 'white',
                   collocations = False,
                   stopwords = stopwords,
                   width = 600, height = 400)

PnPwc_3.generate(text)
PnPwc_3.to_image()
```

```python
PnPwc_3.to_file("PnP_2.png")
```
